```python
# Establish features/target
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

### 2. Develop a multi-layer feed-forward/backpropagation neural network

### 3. Be able to adjust parameters during experiments

### 4. Determine best NN settings

```python
# NN implementation
feature_set = x.values.reshape(6497,12)
labels = y.values.reshape(6497,1)
```

```python
np.random.seed(42)
weights = np.random.rand(12,1)
bias = np.random.rand(1)
lr = 0.05
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
for epoch in range(2000):
    inputs = feature_set

    # feedforward step1
    XW = np.dot(feature_set, weights) + bias

    #feedforward step2
    z = sigmoid(XW)

    # backpropagation step 1
    error = z - labels

    print(error.sum())

    # backpropagation step 2
    dcost_dpred = error
    dpred_dz = sigmoid_der(z)

    z_delta = dcost_dpred * dpred_dz

    inputs = feature_set.T
    weights -= lr * np.dot(inputs, z_delta)

    for num in z_delta:
        bias -= lr * num
```

```python
# Test by trying to predict the value of a single instance
# Use first data entry
single_point = np.array([7.4,0.70,0.00,1.9,0.076,11.0,34.0,0.9978,3.51,0.56,9.4,5])
result = sigmoid(np.dot(single_point, weights) + bias)

# Should be 1
print(result)
```

#### Our from scratch model appears very accurate - let's test our results with a well known NN package

```python
# Split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = .2)
```

```python
# Train NN
from sklearn.neural_network import MLPClassifier

# Create 3 layers of 10 nodes each
mlp = MLPClassifier(hidden_layer_sizes = (10, 10, 10), max_iter = 100)

# Fit model
mlp.fit(x_train, y_train.values.ravel())

# Make predictions
pred = mlp.predict(x_test)
```

```python
# Evaluate the model
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))
```
